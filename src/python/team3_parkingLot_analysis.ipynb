```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, DECIMAL, Date, Float, Boolean, Table

from sqlalchemy.orm import DeclarativeBase
from sqlalchemy.orm import mapped_column


# class Base(DeclarativeBase):
#     pass
Base = declarative_base()


class Park(Base):
    __tablename__  = 'tb_park'

    Seq           = Column(Integer, autoincrement=True, primary_key=True)
    Name          = Column(String(64))
    Type          = Column(String(8))
    Category      = Column(String(8))
    Address       = Column(String(256))
    Spaces        = Column(Integer)
    Latitude      = Column(Float)
    Longitude     = Column(Float)
    DisabledSpace = Column(String(1))
    DataDate      = Column(Date)
    Sido          = Column(String(32))
    Gugun         = Column(String(16))

base = Base()
base.metadata.drop_all(engine)
base.metadata.create_all(engine)
```
