```python
gs_sido_pie = pd.read_csv('./gs_sido.csv', header=None)
```

```python
gs_sido_pie.columns = ['index','frchcd', 'frchnm', 'sido', 'gugun', 'service', 'contact']
```

```python
# sido 열을 기준으로 그룹화, 합계 연산 
gs_sido_pie_sido = gs_sido_pie.groupby('sido').sum(numeric_only=True)
gs_sido_pie_sido.iloc[1:,:].head(20)
```
